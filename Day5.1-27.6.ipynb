# Data Analysis

```python
df.pivot_table(index='class',values='pl') #mean petal length of all types
```

* values=pl ->calc mean based on column pl for all sub table
* index=class -> break table into sub tables based on class values

```python
df.pivot_table(index='class',values=['pl','pw'],aggfunc=["min","median","max"]) 
# aggfunc -> describes func to perform, default func is mean 
```

* Data indicates, Setosa is really small. Virginica is bigger than versicolor, both are relatively big
