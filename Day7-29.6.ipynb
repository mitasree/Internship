```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sdf=pd.read_csv('StudentsPerformance.csv')
newcol=['gender','race','pedu','lunch','prep','math','read','writing']
sdf.columns=newcol
```

# CONTINUOUS DATA

<h3> Math vs Read
<br> Math: continuous
<br> Read: continuous
<br>Hue has to be either categorical or discrete,to groupify. Lesser the no. of unique values, the better

```python
sns.scatterplot(x="math",y="read",hue="prep",data=sdf)
plt.grid(True)
plt.title("Math vs Read")
```

* if no certain conclusion is available then donot present

```python
sns.scatterplot(x="math",y="read",hue="gender",data=sdf)# hue has to be either categorical or discrete
plt.grid(True)
plt.title("Math vs Read")
```

```python
sdf.pivot_table(index="gender",values=["read","math"])
```

* conclusion->for a particular maths marks, females are during better in reading 

```python
sns.scatterplot(x="math",y="read",hue="pedu",data=sdf)# hue has to be either categorical or discrete
plt.grid(True)
plt.title("Math vs Read")
```

* graph of no use

```python
#used to make seperate graphs of a particular data
g=sns.FacetGrid(data=sdf,col="pedu")
#number of unique values of col, facetgrid creates that many no of graph spaces->here 6
g.map(sns.scatterplot,"math","read")
#map used to print a particular type-> here scatterplot,with x and y axis specified
```

```python
g=sns.FacetGrid(data=sdf,row="pedu",col="lunch")
#2 attributes taken
g.map(sns.scatterplot,"math","read")
```

# CATEGORICAL DATA

```python
sns.countplot(x="gender",data=sdf)#frequency count
plt.grid(True)
```

```python
sns.countplot(x="pedu",data=sdf)#frequency count
plt.grid(True)
```

* to overcome rotation the following can be done -->

```python
sns.countplot(x="pedu",data=sdf)#frequency count
plt.grid(True)
plt.xticks(rotation=90)
```

```python
sns.countplot(x="pedu",data=sdf)#frequency count
plt.grid(True)
plt.xticks(rotation=45)
```

```python
sns.countplot(y="pedu",data=sdf)#frequency count
plt.grid(True)
```

```python
sns.countplot(x="prep",hue="gender",data=sdf)#frequency count
plt.grid(True)
```

```python
pd.crosstab(index=sdf["prep"],columns=sdf["gender"]) #only to be used with categorical data
```

* pivot table -> summarising measurement column based on some categorical column
* crosstab -> both categorical columns

```python
pd.crosstab(index=sdf["prep"],columns=sdf["gender"],margins=True)#summation of all
```

```python
pd.crosstab(index=sdf["prep"],columns=sdf["gender"],normalize="index")#percentage calculation
```

```python
pd.crosstab(index=sdf["race"],columns=sdf["gender"],normalize="index")
```

```python
pd.crosstab(index=sdf["pedu"],columns=sdf["race"],normalize="index")
```

```python
sns.heatmap(pd.crosstab(index=sdf["pedu"],columns=sdf["race"],normalize="index"),annot=True)
```
