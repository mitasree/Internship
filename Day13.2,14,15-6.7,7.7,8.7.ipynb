```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import model_selection
from sklearn import linear_model
from sklearn import metrics
from sklearn import preprocessing
from sklearn import tree
from sklearn import feature_selection
from sklearn import ensemble #for RandomForest
```

```python
df=pd.read_csv('IRIS.csv')
df["type"]=df["species"].replace({"Iris-setosa":1,"Iris-versicolor":2,"Iris-virginica":3})
df
```

```python
df['type'].value_counts()
```

```python
X=df.drop(["species","type"],axis=1)
y=df["type"]
Xtrain,Xtest,ytrain,ytest=model_selection.train_test_split(X,y,test_size=.15,random_state=42)
```

```python
treemodel=tree.DecisionTreeClassifier(random_state=42)
treemodel.fit(Xtrain,ytrain) #binary tree model
predtrain=treemodel.predict(Xtrain)
predtest=treemodel.predict(Xtest)
```

```python
from sklearn import metrics

def printscores(act, pred):
    print("accuracy :", metrics.accuracy_score(act, pred))
    print("recall (macro):", metrics.recall_score(act, pred, average="macro"))
    print("precision (macro):", metrics.precision_score(act, pred, average="macro"))
    print("f1 (macro):", metrics.f1_score(act, pred, average="macro"))
```

```python
print("TRAINING MERICS:")
printscores(ytrain,predtrain)
print("==================")
print("TEST METRICS:")
printscores(ytest,predtest)
```

<h3> PIMA DIABETICS

```python
df=pd.read_csv('diabetes.csv')
df.info()
```

```python
df=df.drop(["SkinThickness","Insulin"],axis=1)
```

```python
df=df[df['Pregnancies']!=0]
df=df[df['BloodPressure']!=0]
df=df[df['BMI']!=0]
X=df.drop("Outcome",axis=1)
y=df["Outcome"]

Xtrain,Xtest,ytrain,ytest=model_selection.train_test_split(X,y,test_size=.25,random_state=42)
```

```python
model=tree.DecisionTreeClassifier(random_state=42)
model.fit(Xtrain,ytrain)
predtrain=model.predict(Xtrain)
predtest=model.predict(Xtest)
print("TRAINING MERICS:")
printscores(ytrain,predtrain)
print("==================")
print("TEST METRICS:")
printscores(ytest,predtest)
```

<h3> Recursive Feature Elimination using LR

```python
#Creating a model using most important columns.We are using Logistic Regression to find the model.RFE class used for dimensionality reduction.
```

```python
model=linear_model.LogisticRegression()
rfeobj=feature_selection.RFE(estimator=model,n_features_to_select=4) #RFE->class under feature_selelction
rfeobj.fit(Xtrain,ytrain)
```

```python
rfeobj.support_
```

```python
Xtrain.columns
```

```python
print("SIGNIFICANT COLUMNS :",Xtrain.columns[rfeobj.support_])
```

```python
arr=rfeobj.transform(Xtrain) #returns a numpy array that needs to be changed back to dataframe
Xtrain1=pd.DataFrame(arr,columns=Xtrain.columns[rfeobj.support_],index=Xtrain.index)
#updating columns according to the new columns from rfeobj, index updated according to the Xtrain index so that the selected index is recovered
Xtest1=pd.DataFrame(rfeobj.transform(Xtest),columns=Xtest.columns[rfeobj.support_],index=Xtest.index)
```

```python
Xtrain1[:10]
```

<h3> Recursive Feature Eliminstion using DT

```python
#using Dec Tree model is formed now
model=tree.DecisionTreeClassifier(random_state=42)
rfeobj=feature_selection.RFE(estimator=model,n_features_to_select=4) #RFE->class under feature_selelction
rfeobj.fit(Xtrain,ytrain)
print("SIGNIFICANT COLUMNS :",Xtrain.columns[rfeobj.support_])
arr=rfeobj.transform(Xtrain) #returns a numpy array that needs to be changed back to dataframe
Xtrain1=pd.DataFrame(arr,columns=Xtrain.columns[rfeobj.support_],index=Xtrain.index)
#updating columns according to the new columns from rfeobj, index updated according to the Xtrain index so that the selected index is recovered
Xtest1=pd.DataFrame(rfeobj.transform(Xtest),columns=Xtest.columns[rfeobj.support_],index=Xtest.index)
```

```python
#using Xtrain1 model
model=tree.DecisionTreeClassifier(random_state=42)
model.fit(Xtrain1,ytrain) 
predtrain=model.predict(Xtrain1)
predtest=model.predict(Xtest1)
print("TRAINING MERICS:")
printscores(ytrain,predtrain)
print("==================")
print("TEST METRICS:")
printscores(ytest,predtest)
```

```python
model=tree.DecisionTreeClassifier(random_state=42,max_depth=6)
model.fit(Xtrain1,ytrain) 
predtrain=model.predict(Xtrain1)
predtest=model.predict(Xtest1)
print("TRAINING MERICS:")
printscores(ytrain,predtrain)
print("==================")
print("TEST METRICS:")
printscores(ytest,predtest)
```

```python
model=tree.DecisionTreeClassifier(random_state=42,min_samples_split=20)
model.fit(Xtrain1,ytrain) 
predtrain=model.predict(Xtrain1)
predtest=model.predict(Xtest1)
print("TRAINING MERICS:")
printscores(ytrain,predtrain)
print("==================")
print("TEST METRICS:")
printscores(ytest,predtest)
```

<h2>Hyper Parameter Tuning

```python
#GridsearchCV-> class used to tune the parameters
```

```python
#Step 1: Create dictionary
grid={ "max_depth":[8,9,10,11,12] }  # param_grid, key== hyperparameter name
model=tree.DecisionTreeClassifier(random_state=42) #estimator
gobj=model_selection.GridSearchCV(estimator=model,param_grid=grid,cv=5,scoring="recall",n_jobs=-1,return_train_score=True)
gobj.fit(Xtrain,ytrain) ## VERY INTENSIVE AND CPU CONSUMING TASK BEGINS
```

```python
print(gobj.best_params_) # we see it works best on 8,maybe it will have better with previous data, so we check with 3-8
```

```python
print(gobj.best_score_)
```

```python
bestmodel=gobj.best_estimator_ # directly selects the best parameter sets after tuning and creates new best model
```

```python
predtest=bestmodel.predict(Xtest)
predtrain=bestmodel.predict(Xtrain)
```

```python
printscores(ytest,predtest)
```

```python
printscores(ytrain,predtrain)
```
